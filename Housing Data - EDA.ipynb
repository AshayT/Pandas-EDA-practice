```python
housing_sub = housing.loc[housing['LotArea'] <housing['LotArea'].quantile(0.965)]
```

```python
sns.regplot( x= 'LotArea',y  = 'SalePrice', data = housing_sub)
```

```python
col_sf = [col for col in housing.columns if 'SF' in col]
```

```python
col_sf
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=3, ncols =3, figsize = (20,20))
for i in range(len(col_sf)):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.regplot(x = col_sf[i], y = 'SalePrice', data = housing_sub, ax = ax)
```

```python
for i in range(len(col_sf)):
    row = i//3
    col = i%3
    axes = (row,col)
    sns.regplot(x = col_sf[i], ax = ax, y = 'SalePrice', data = housing_sub) #this alone does not give any output
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,10))
plot = sns.boxplot(x='Exterior1st', y = 'SalePrice', data = housing_sub)
plot.set_xticklabels(plot.get_xticklabels(), rotation = 50);
```
