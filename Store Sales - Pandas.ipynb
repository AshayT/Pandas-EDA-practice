```python
import numpy as np
def bonusf(column): #elibility for bonus based on Jan (if sales >10)
    eligibility = np.where(column>10, 'Eligible', 'Not eligible')
    return eligibility
```

```python
store['Bonus'] = bonusf(store['Jan'])
```

```python
store
```

```python
new_bonus =  store.apply(bonusf) #so the datatypes should be mentioned
```

```python
new_bonus =  store.select_dtypes('int64', 'float64').apply(bonusf)
```

```python
new_bonus
```

```python
new_frame = pd.concat([store, new_bonus])
```
